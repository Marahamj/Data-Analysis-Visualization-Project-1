```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
df_all = pd.read_csv('insurance.csv')
df=df_all.head(500)

print(df.info())

df.describe()
```

```python
print(df.isnull().sum())
#إزالة الصفوف المكررة
df= df.drop_duplicates()
print(f"Minimum age: {df['age'].min()}")
```

```python
df['sex'] = df['sex'].astype(str).str.strip().str.lower()
df['smoker'] = df['smoker'].astype(str).str.strip().str.lower()


df['smoker'] = df['smoker'].map({'yes': 1, 'no': 0})
df['sex'] = df['sex'].map({'female': 1, 'male': 0})
print("\n--- Data after Encoding ---")
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols_to_scale = ['bmi', 'charges']
df[[ 'bmi_scaled', 'charges_scaled']] = scaler.fit_transform(df[cols_to_scale])
print(df[['sex', 'smoker','bmi_scaled']].head())
```

3.
Data Visualization

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bmi_scaled', y='charges_scaled', hue='smoker', palette='coolwarm')

sns.regplot(data=df, x='bmi_scaled', y='charges_scaled', scatter=False, color='black')

plt.title('Relationship between BMI and Charges (Scaled)')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
# حساب الارتباط بين الأعمدة الرقمية والمحولة
corr = df[['sex', 'smoker',  'bmi_scaled', 'charges_scaled']].corr()
# رسم الخريطة الحرارية مع إظهار الأرقام (annot=True)
sns.heatmap(corr, annot=True)

plt.title('Heatmap of Feature Correlation')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# رسم منحنى الكثافة
sns.kdeplot(data=df, x='charges_scaled', hue='smoker')

plt.title('Distribution of Charges by Smoking Status')
plt.show()
```
